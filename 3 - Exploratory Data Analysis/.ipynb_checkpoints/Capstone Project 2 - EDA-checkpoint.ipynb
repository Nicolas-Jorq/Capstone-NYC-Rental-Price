## 4. Scatterplots of numeric features against Rent Price

```python
def scatterplots(columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x = col, y = 'Cost', data=df, alpha=0.5)
        ax.set(xlabel=col, ylabel='Rent Price')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
features = [x for x in df.columns if x not in ['Unit_Type','Address', 'Home_Type', 'Neighborhood', 'boro_names', 'Postal_Code']]
```

```python
scatterplots(features, ncol=4, figsize=(15, 15))
```

## 5. Visualizing High Dimensional Data

## Zillow DataFrame

### 5.1 Scaling the Data

```python
# Removed NULL Values so that the data could scale
Numerical_Data = df.drop(['Unit_Type','Address', 'Home_Type', 'Neighborhood', 'boro_names', 'Postal_Code','Area_SqFt'], axis=1).dropna()
Numerical_Data.describe()
```

```python
df_scaled = scale(Numerical_Data)
df_scaled = pd.DataFrame(df_scaled, columns=Numerical_Data.columns)
df_scaled.head()
```

```python
df_scaled.info()
```

```python
df_scaled.std(ddof=0)
```

Our data has scale appropriately as the standard deviation is 1

### 5.2 PCA Transformation

```python
rent_pca = PCA().fit(df_scaled)
```

```python
plt.subplots(figsize=(10, 6))
plt.plot(rent_pca.explained_variance_ratio_.cumsum())
plt.xlabel('Component #')
plt.ylabel('Cumulative ratio variance')
plt.title('Cumulative variance ratio explained by PCA components for Rent Price summary statistics');
```

The first three components seem to account for over 70% of the variance, and the first five for over 80%.

```python
rent2_pca = rent_pca.transform(df_scaled)
```

```python
rent2_pca.shape
```

### 5.3 Exploring PCA Transformation

```python
x = rent2_pca[:, 0]
y = rent2_pca[:, 1]
pc_var = 100 * rent_pca.explained_variance_ratio_.cumsum()[1]
```

```python
plt.subplots(figsize=(10,8))
plt.scatter(x=x, y=y)
plt.xlabel('First component')
plt.ylabel('Second component')
plt.title(f'PCA, {pc_var:.1f}% variance explained')
```

```python
df.reset_index(inplace=True)
```

```python
pca_df = pd.DataFrame({'PC1': rent2_pca[:, 0], 'PC2': rent2_pca[:, 1]})
pca_df = pd.concat([pca_df, df.median_rent], axis=1)
pca_df['boro_names'] = df['boro_names']
pca_df.head()
```

```python
pca_df.shape
```

```python
pca_df[pca_df.isnull().any(axis=1)]
```

```python
Neighborhood[Neighborhood.median_rent == 1362.07]
```

Interisting to note that missing values all pertain to the Rockaways. 

```python
pca_df.dropna()
```

```python
pca_df['Quartile'] = pd.qcut(pca_df.median_rent, q=5, precision=1)
pca_df.head()
```

```python
x = pca_df.PC1
y = pca_df.PC2

plt.subplots(figsize=(12, 10))
# Note the argument below to make sure we get the colours in the ascending
# order we intuitively expect!
sns.scatterplot(x='PC1', y='PC2', size='median_rent', hue='boro_names', 
                data=pca_df)
plt.title(f'Zillow Rent summary PCA, {pc_var:.1f}% variance explained');
```

## Neighborhood DataFrame

### 5.1 Scaling the Data

```python
Neighborhood.head()
```

```python
# REMOVING ALL CATEGORICAL DATA
Numerical_Data2 = Neighborhood.drop(['Neighborhood','boro_names'], axis=1)
```

```python
# SCALING THE DATA
df_scaled2 = scale(Numerical_Data2)
df_scaled2 = pd.DataFrame(df_scaled2, columns=Numerical_Data2.columns)
df_scaled2.head()
```

```python
# PCA TRANSFROMATION
neighborhood_pca = PCA().fit(df_scaled2)
plt.subplots(figsize=(10, 6))
plt.plot(neighborhood_pca.explained_variance_ratio_.cumsum())
plt.xlabel('Component #')
plt.ylabel('Cumulative ratio variance')
plt.title('Cumulative variance ratio explained by PCA components for Rent Price summary statistics');
```

```python
neighborhood2_pca = neighborhood_pca.transform(df_scaled2)
```

```python
Neighborhoods = Neighborhood.Neighborhood
pca_df2 = pd.DataFrame({'PC1': neighborhood2_pca[:, 0], 'PC2': neighborhood2_pca[:, 1]})
pca_df2.set_index(Neighborhoods, inplace=True)
pca_df2['boro_names'] = Neighborhood['boro_names'].values
pca_df2['median_rent'] = Neighborhood['median_rent'].values
pca_df2.head()
```

```python
x = pca_df2.PC1
y = pca_df2.PC2
pc_var2 = 100 * neighborhood_pca.explained_variance_ratio_.cumsum()[1]

plt.subplots(figsize=(10, 8))
# Note the argument below to make sure we get the colours in the ascending
# order we intuitively expect!
sns.scatterplot(x='PC1', y='PC2', hue='boro_names', size='median_rent', sizes = (100,250),
                data=pca_df2)
plt.title(f'Zillow Rent summary PCA, {pc_var2:.1f}% variance explained');
```
